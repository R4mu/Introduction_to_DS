```python
def one_hot_encode(problem2_X_train, problem2_X_test):


    # Sex is categorical! Encode it
    s = (problem2_X_train.dtypes== 'object')
    object_cols = list(s[s].index)
    object_cols

    from sklearn.preprocessing import OneHotEncoder

    # Apply one-hot encoder to each column with categorical data
    ohe = OneHotEncoder(sparse_output=False)

    # when applying fit_transform using OHE, it returns a numpy array. Remember to transform it into a data frame
    OH_cols_train = pd.DataFrame(ohe.fit_transform(problem2_X_train[object_cols]))
    OH_cols_test = pd.DataFrame(ohe.fit_transform(problem2_X_test[object_cols]))

    # get the proper column names
    OH_cols_train.columns = ohe.get_feature_names_out(object_cols)
    OH_cols_test.columns = ohe.get_feature_names_out(object_cols)

    # the indices were removed by OHE, put them back
    OH_cols_train.index =problem2_X_train.index
    OH_cols_test.index = problem2_X_test.index

    # Remove categorical columns
    num_X_train = problem2_X_train.drop(object_cols, axis=1)
    num_X_test = problem2_X_test.drop(object_cols, axis=1)

    # Concat the categorical and numerical features back together
    problem2_X_train = pd.concat([num_X_train, OH_cols_train], axis=1)
    problem2_X_test = pd.concat([num_X_test, OH_cols_test], axis=1)

    return problem2_X_train, problem2_X_test



problem2_X_train,problem2_X_test = one_hot_encode(problem2_X_train,problem2_X_test)
```

```python
#Part 1: Load the Data and Define Features and Target

import pandas as pd

# Load the dataset into a DataFrame
problem2_df = pd.read_csv("data/abalone.csv")

# Features are all columns except "Rings" (which is the target)
problem2_features = [
    "Length", "Diameter", "Height", "WholeWeight",
    "ShuckedWeight", "VisceraWeight", "ShellWeight"
]

# Target column is "Rings"
problem2_target = "Rings"


#Part 2: Split the Data into Train and Test
from sklearn.model_selection import train_test_split

# Split the dataset into features (X) and target (y)
X = problem2_df[problem2_features]  #X = problem2_df[problem2_features].values
y = problem2_df[problem2_target]  #y = problem2_df[problem2_target].values

# Split into training and testing sets (80% train, 20% test)
problem2_X_train, problem2_X_test, problem2_y_train, problem2_y_test = train_test_split(
    X, y, train_size=0.8, random_state=42
)


#Part 3: Train the Model
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression model
problem2_model = LinearRegression()

# Train the model on the training data
problem2_model.fit(problem2_X_train, problem2_y_train)

#Part 4: Evaluate the Model and Plot the Empirical Distribution Function (EDF)
#Compute Mean Absolute Error (MAE):
from sklearn.metrics import mean_absolute_error

# Predict on the test set
predictions = problem2_model.predict(problem2_X_test)

# Compute Mean Absolute Error
problem2_mae = mean_absolute_error(problem2_y_test, predictions)
print("Mean Absolute Error (MAE):", problem2_mae)

#Plot EDF of Residuals with Confidence Bands:
import numpy as np
import matplotlib.pyplot as plt

# Residuals
residuals = problem2_y_test - predictions

# Function to create EDF
def makeEDF(data):
    sorted_data = np.sort(data)
    cdf = np.arange(1, len(data) + 1) / len(data)
    return sorted_data, cdf

# Function to plot EDF with DKW confidence bands
def plotEDF(data, alpha=0.05):
    sorted_data, cdf = makeEDF(data)
    n = len(data)
    epsilon = np.sqrt(np.log(2 / alpha) / (2 * n))  # DKW inequality

    plt.figure(figsize=(8, 6))
    plt.step(sorted_data, cdf, where="post", label="Empirical CDF")
    plt.fill_between(
        sorted_data,
        np.maximum(cdf - epsilon, 0),
        np.minimum(cdf + epsilon, 1),
        color="blue", alpha=0.2, label="95% Confidence Bands"
    )
    plt.xlabel("Residual")
    plt.ylabel("Cumulative Probability")
    plt.title("Empirical Distribution Function of Residuals")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot EDF with confidence bands
plotEDF(residuals)

#Part 5: Scatter Plot of Predictions vs True Values
# Scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(predictions, problem2_y_test, alpha=0.6, edgecolor="k")
plt.plot([problem2_y_test.min(), problem2_y_test.max()],
         [problem2_y_test.min(), problem2_y_test.max()],
         color="red", linestyle="--", label="Perfect Prediction")
plt.xlabel("Predicted Rings")
plt.ylabel("True Rings")
plt.title("Scatter Plot: Predicted vs True Rings")
plt.legend()
plt.grid(True)
plt.show()



```

```python
class PoissonRegression(object):
    def __init__(self):
        self.coeffs = None
        self.result = None

    def fit(self, X, Y):
        import numpy as np
        from scipy import optimize

        # Define the objective/cost/loss function we want to minimize
        def loss(coeffs):
            # The parameter lambda for the given X and the proposed values of the coefficients
            lam = np.exp(np.dot(X, coeffs[:-1]) + coeffs[-1])
            # Compute the negative log-likelihood
            return -np.sum(Y * np.log(lam) - lam)

        # Use the loss above with an optimization method from scipy to find the coefficients
        initial_arguments = np.zeros(shape=X.shape[1] + 1)  # Initial guess as 0
        self.result = optimize.minimize(loss, initial_arguments, method='cg')
        self.coeffs = self.result.x

    def predict(self, X):
        # Use the trained model to predict Y
        if self.coeffs is not None:
            return np.exp(np.dot(X, self.coeffs[:-1]) + self.coeffs[-1])
```
